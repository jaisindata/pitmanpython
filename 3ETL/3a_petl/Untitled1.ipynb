```python
emp = etl.fromtext('./ExtractingandLoadingDelimitedAndTextFiles/emp_details.txt',header=['details'])
```

```python
emp.display()
```

```python
emptable = emp.capture('details','(.*),(.*),(.*)$',['id','name','dept'])
```

```python
emptable
```

```python
template = '''|->
|{id}
|{name}
|{dept}'''
```

```python
epilogue = '|}'
```

```python
etl.totext(emptable,'./datasets/emp.txt',template=template,epilogue=epilogue)
```

```python
print(open('./datasets/emp.txt').read())
```

```python
template = '$$$$$$$\n''{id},{name},{dept}\n'
```

```python
prologue = '%%%%%%%%%%%Start%%%%%%%%%%\n'
epilogue='%%%%%%%%%%%%%End%%%%%%%%'
```

```python
etl.totext(emptable,'./datasets/emp.txt',template=template,epilogue=epilogue,prologue=prologue)
```

```python
print(open('./datasets/emp.txt').read())
```
