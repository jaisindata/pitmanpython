```python
import pandas as pd
import openpyxl
```

```python
data = {
    'country' : ['India','England','France'],
    'capital' : ['NewDelhi','London','Paris'],
    'currency' : ['Rupee','Pound','Franc']
}
```

```python
info_df = pd.DataFrame(data)
```

```python
for row in info_df.itertuples():
    print(row)
```

```python
for row_index,row in info_df.iterrows():
    print(row_index)
    print(row)
```

```python
for key,value in info_df.items():
    print(key)
    print(f"{value}\n")
```

```python
info_df.to_excel('datasets/info_df.xlsx',index=False)
```

```python
info_df.to_csv('datasets/info_df.txt',index=False)
```

```python
info_df.to_json('datasets/info_df.json')
```

```python
info_df.to_html('datasets/info_df.html',index=False)
```
