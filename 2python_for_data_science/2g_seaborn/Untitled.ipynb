```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pollution = pd.read_csv('datasets/UFOPOLLUTANTS.csv').head(1000)
```

```python
pollution['NO2.AQI'].head(10)
```

```python
pollution.describe()
```

```python
sns.distplot(pollution['NO2.AQI'])
plt.show()
```

```python
dist = sns.distplot(pollution['NO2.AQI'],color='green',vertical=True)
dist.set_title = "Pollution NO2"
dist.set_xlabel = "Pollution level"
dist.set_ylabel = "Density"
plt.show()
```

```python
dist = sns.distplot(pollution['NO2.AQI'],kde_kws = {"color":"b","lw":2,"label":"KDE"},hist_kws={"color":"g","label":"hist"})
dist.set_title = "Pollution NO2"
dist.set_xlabel = "Pollution level"
dist.set_ylabel = "Density"
plt.show()
```

```python
sns.set(color_codes = True)
plot = sns.kdeplot(pollution['NO2.AQI'],color='b',fill=True)
```

```python
"""help(sns.jointplot)"""
jointplot = sns.jointplot(data=pollution,x='NO2.Mean',y='CO.Mean',kind='reg')
plt.show()
```

```python
sns.pairplot(pollution,height=5,vars=['NO2.Mean','CO.Mean','O3.Mean','SO2.Mean'],hue='year')
```

```python
pollutants_mean = pollution[['NO2.Mean','CO.Mean','O3.Mean','SO2.Mean']]
```

```python
corrmatrix = pollutants_mean.corr()
plt.subplots(figsize = (7,7))
sns.heatmap(corrmatrix,vmax=.8,square=True,annot=True,fmt='.2f',cmap='YlGnBu')
plt.show()
```

```python
sns.lmplot(pollution,x='NO2.Mean',y='CO.Mean',col='year')
plt.show()
```
