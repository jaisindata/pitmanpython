```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
emp_data = pd.read_csv('datasets/Absenteeism_at_work.csv')
```

```python
plt.subplots(figsize=(12,6))
sns.distplot(emp_data['Reason for absence'],bins=29,kde=False,rug=True)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.distplot(emp_data['Month of absence'],bins=12,kde=False,rug=True)
plt.show()
```

```python
sns.jointplot(x='Day of the week',y='Reason for absence',data = emp_data,height=6)
plt.show()
```

```python
sns.jointplot(x='Seasons',y='Reason for absence',data = emp_data,height=6)
plt.show()
```

```python
sns.axes_style()
```

```python
car_data = pd.read_csv('datasets/mpg.csv')
```

```python
plt.subplots(figsize=(12,6))
sns.boxplot(x='cylinders',y='horsepower',data= car_data)
plt.show()
```

```python
plt.subplots(figsize=(12,6))
sns.violinplot(x='cylinders',y='horsepower',data= car_data,scale='count')
plt.show()
```
